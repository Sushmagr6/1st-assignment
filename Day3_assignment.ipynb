```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import randn, randint
%matplotlib inline
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, style='event', size='region')
plt.show()
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, style='event', size='region')
plt.show()
```

```python
sns.catplot(x='timepoint', y='signal', kind='box', data=fmri)
plt.show()
```
